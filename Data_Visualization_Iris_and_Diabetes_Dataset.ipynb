#Introduction

The Iris dataset and Diabetes dataset are two widely used datasets in data science and machine learning, each serving a different analytical purpose.

The Iris dataset is a well-known dataset in classification tasks, containing 150 samples of three different species of iris flowers (Setosa, Versicolor, and Virginica). It includes four key features—sepal length, sepal width, petal length, and petal width—which help distinguish between species. This dataset is commonly used to demonstrate clustering and classification algorithms.

The Diabetes dataset is crucial for medical and health-related analysis. It includes patient health indicators such as Age, BMI, Glucose levels, Blood Pressure, and Insulin levels, which are used to assess diabetes risk and progression. By exploring relationships between these features, we can identify key risk factors and patterns that contribute to diabetes development.

```python
 # Q1: Distribution of sepal lengths and widths among species

plt.figure(figsize=(12, 6))
sns.histplot(df1, x ='sepal_length', hue='species', kde=True, bins=20)
plt.title('Distribution of Sepal Length by Species')
plt.show()
```

There is a noticeable overlap in sepal lengths between Versicolor and Virginica, making it harder to distinguish them based on this feature alone. However, Setosa is distinctly separate due to its smaller sepal lengths.
This suggests that sepal length is a useful feature for distinguishing Setosa, but additional features like petal measurements may be needed to differentiate Versicolor and Virginica.

```python
plt.figure(figsize=(12, 6))
sns.histplot(df1, x='sepal_width', hue='species', kde=True, bins=20)
plt.title('Distribution of Sepal Width by Species')
plt.show()
```

Setosa has distinctly larger sepal widths, making this feature effective for separating it from Versicolor and Virginica.
Versicolor and Virginica overlap significantly in sepal width, indicating that this feature alone may not effectively differentiate these two species.
Combining sepal width with other features (like petal dimensions) might provide better separation for all three species.

```python
 # Q2: Petal length variation across species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_length', data=df1)
plt.title('Petal Length Variation Across Species')
plt.show()
```

Virginica has distinctly larger spetal lengths, making this feature effective for separating it from Versicolor and Setosa.
No overlapping is found wrt petal length between Versicolor and Setosa.

```python
 # Q3: Relationship between sepal length and petal length
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df1)
plt.title('Sepal Length vs Petal Length')
plt.show()
```

Each species of dataset shows distinct degree of relationship between petal length and sepal length which makes the classification easier with these features

```python
 # Q4: Pairwise relationships between features
sns.pairplot(df1, hue='species')
plt.show()
```

Sepal Length vs. Sepal Width: There is some separation between species, but overlap exists, especially between Setosa and the other species. Setosa tends to have a lower sepal length and higher sepal width.

Petal Length vs. Petal Width: This pair shows the most clear distinction between species. Setosa has significantly smaller petals, while Versicolor and Virginica form somewhat distinct clusters.

Sepal Length vs. Petal Length: A strong positive correlation is observed—longer sepals tend to be associated with longer petals.

Sepal Width vs. Petal Width: The relationship is less clear, but Setosa remains distinct from the other specie

```python
 # Q5: Distinguishing species based on sepal and petal measurements
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df1)
plt.title('Petal Length vs Petal Width')
plt.show()
```

Petal-based measurements (length & width) are the most reliable for distinguishing species. Setosa is well-separated, while Versicolor and Virginica show some overlap but can still be classified using petal features.

```python
# Q6: Distribution of Age and BMI among diabetic patients
plt.figure(figsize=(12, 6))
sns.histplot(df2['Age'], bins=30, kde=True)
plt.title('Age Distribution in Diabetes Dataset')
plt.show()
```

The age distribution show a right-skewed pattern, indicating that diabetes is more prevalent in middle-aged and older individuals.

If most diabetic patients fall within the 40-60 age range, it suggests that diabetes risk increases with age.

Younger individuals (<30) might have fewer cases, indicating that diabetes is less common at an early age.

```python
plt.figure(figsize=(12, 6))
sns.histplot(df2['BMI'], bins=30, kde=True)
plt.title('BMI Distribution in Diabetes Dataset')
plt.show()
```

The BMI histogram may indicate that diabetic patients tend to have higher BMI values.

If a majority of cases fall in the 25-35 BMI range, it suggests that overweight and obesity are strong risk factors for diabetes.

A right-skewed distribution would indicate that extremely high BMI values (e.g., >40) are present in some cases, highlighting severe obesity in some diabetic patients.

```python
# Q7: Blood pressure levels across age groups
plt.figure(figsize=(12, 6))
sns.boxplot(x=pd.cut(df2['Age'], bins=[20, 30, 40, 50, 60, 70, 80]), y='BloodPressure', data=df2)
plt.title('Blood Pressure Across Age Groups')
plt.show()
```

Blood pressure rises with age, making age an important factor in hypertension and diabetes risk.

If the boxplot shows a high number of outliers in older age groups, it may indicate high-risk individuals requiring closer monitoring.

A steep increase in median blood pressure between middle-aged and older groups may highlight the importance of early intervention strategies to manage hypertension.

```python
# Q8: Correlation between BMI and Glucose levels
plt.figure(figsize=(8, 6))
sns.scatterplot(x='BMI', y='Glucose', data=df2)
plt.title('BMI vs Glucose Levels')
plt.show()
print("Correlation between BMI and Glucose:", df2[['BMI', 'Glucose']].corr())
```

BMI and Glucose levels often show a moderate positive correlation (typically between 0.3 and 0.6), indicating that individuals with higher BMI are more likely to have higher glucose levels.

If the correlation is strong (above 0.6), it suggests that BMI is a significant factor influencing glucose levels, possibly due to insulin resistance in overweight individuals.

If the correlation is weak (below 0.3), it implies that BMI alone may not be a strong predictor of glucose levels, and other factors (e.g., diet, genetics, activity levels) play a major role.

A higher BMI is often associated with higher glucose levels, supporting the link between obesity and diabetes risk.

A strong correlation suggests weight management strategies could help regulate glucose levels.

If correlation varies across subgroups (e.g., diabetic vs. non-diabetic), it could indicate different metabolic responses to BMI changes.

```python
# Q9: Diabetes progression vs. patient age
plt.figure(figsize=(12, 6))
sns.boxplot(x=pd.cut(df2['Age'], bins=[20, 30, 40, 50, 60, 70, 80]), y='Outcome', data=df2)
plt.title('Diabetes Outcome Across Age Groups')
plt.show()
```

If age correlates strongly with diabetes progression, then early intervention strategies should focus on middle-aged individuals to slow disease progression.

If no strong correlation exists, then other factors (e.g., BMI, lifestyle, diet) should be analyzed to determine the main drivers of diabetes severity.

Further analysis with a regression model could quantify the exact impact of age on diabetes progression.


```python
# Q10: Insulin levels across different age groups
plt.figure(figsize=(12, 6))
sns.boxplot(x=pd.cut(df2['Age'], bins=[20, 30, 40, 50, 60, 70, 80]), y='Insulin', data=df2)
plt.title('Insulin Levels Across Age Groups')
plt.show()
```

If the median insulin levels increase with age, it suggests that older individuals may have higher insulin resistance or require more insulin regulation.

If insulin levels decrease in older age groups, it might indicate pancreatic β-cell dysfunction, where the body produces less insulin over time.

Young Adults (20-30 years):

Likely to show lower insulin levels with a relatively small interquartile range (IQR).
Outliers may indicate early signs of insulin resistance or metabolic issues.

Middle-Aged Groups (30-50 years):

If insulin levels increase, it suggests rising insulin resistance (a common precursor to Type 2 diabetes).
A wider IQR indicates variability in metabolic health within this age group.

Older Adults (50+ years):

If the boxplot shows a drop in median insulin levels, it could signal age-related insulin production decline.
A high number of outliers may indicate that some older individuals rely heavily on insulin therapy.
